```python
# Creating simple visualizations
%matplotlib inline
import csv
import pandas as pd
import matplotlib
matplotlib.style.use('ggplot')
df = pd.read_csv('..\Data Files\Phoenix\Restaurants_Phoenix.csv', header=0)
```

```python
# 1. Counts of avg. star ratings of restaurants 
df.stars.hist()
```

```python
# 2. Geographic locations of restaurants as per latitude and longitude
df.plot.scatter(x='longitude',y='latitude')
```

```python
# 3. Geographically located restaurants wrt review count
df.plot.scatter(x='longitude',y='latitude', s=df['review_count']*0.2)
```

```python
# 4. Finding top categories and plotting a histogram
dict_cats = {}
cats = df.categories
for row in cats:
    row = row[1:-1]
    row=row.replace("'","")
    entry = row.split(',')    
    for item in entry:        
        if item[0] ==' ':
            item = item[1:]
        if item not in dict_cats:
            dict_cats[item] = 1
        else:
            exist_val = dict_cats[item]
            dict_cats[item] = exist_val + 1

new_dict = {}
for item in dict_cats:
    if dict_cats[item] >= 100:
        new_dict[item] = dict_cats[item]

del new_dict['Restaurants']
del new_dict['Food']
print new_dict
import matplotlib.pyplot as plt
plt.bar(range(len(new_dict)), new_dict.values(), align='center')
```
